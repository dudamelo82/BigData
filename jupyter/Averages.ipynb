```python
import random
import matplotlib.pyplot as plt

# Ask how many users to simulate
num_users = int(input("Enter the number of users: "))

# Generate random data
heights = [random.randint(150, 200) for _ in range(num_users)]  # cm
weights = [random.randint(50, 100) for _ in range(num_users)]   # kg
ages = [random.randint(18, 70) for _ in range(num_users)]       # years

# Calculate averages
avg_height = sum(heights) / num_users
avg_weight = sum(weights) / num_users
avg_age = sum(ages) / num_users

# Show averages in text
print(f"\nAverage height: {avg_height:.1f} cm")
print(f"Average weight: {avg_weight:.1f} kg")
print(f"Average age: {avg_age:.1f} years")

# Prepare data for plotting
categories = ["Height (cm)", "Weight (kg)", "Age (years)"]
averages = [avg_height, avg_weight, avg_age]

# Plot bar chart
plt.bar(categories, averages, color=["skyblue", "salmon", "lightgreen"])
plt.title("Average User Statistics")
plt.ylabel("Values")
plt.show()
```
